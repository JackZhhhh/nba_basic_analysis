```python
#Import team stats
from nba_api.stats.static import teams

nba_teams = teams.get_teams()

team_id = input("Enter abbreviation of team of interest: ")

home_team = [team for team in nba_teams if team["abbreviation"] == team_id][0]
team_id_no = home_team["id"]

#Import all games from team
from nba_api.stats.endpoints import leaguegamefinder

#Game Finder of all previous games
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_id_no)

# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]

year = input("Enter interest of year (e.g. if interested in 2025-2026 input 2025: ")
games_currseason = games[games.SEASON_ID.str[-4:] == year]

#Analysis

#Get Averages

avg_pts = games_currseason['PTS'].mean()
avg_fg_pct = games_currseason['FG_PCT'].mean()
avg_reb = games_currseason['REB'].mean()
avg_oreb = games_currseason['OREB'].mean()
avg_stl = games_currseason['STL'].mean()
avg_blk = games_currseason['BLK'].mean()
avg_tov = games_currseason['TOV'].mean()
avg_pm = games_currseason['PLUS_MINUS'].mean()

#Get other team

other_id = input("Enter abbreviation of team of interest: ")

#AVG against a team
games_against = games_currseason[games_currseason.MATCHUP.str.contains(other_id)]

avg_pts_against = games_against['PTS'].mean()
avg_fg_pct_against = games_against['FG_PCT'].mean()
avg_reb_against = games_against['REB'].mean()
avg_oreb_against = games_against['OREB'].mean()
avg_stl_against = games_against['STL'].mean()
avg_blk_against = games_against['BLK'].mean()
avg_tov_against = games_against['TOV'].mean()
avg_pm_against = games_against['PLUS_MINUS'].mean()

#Analysis

team_name = other_id
print(f"\n=== Matchup Analysis vs {team_name} ===\n")

# --- OFFENSE ---
print("OFFENSE:")

if avg_pts_against > avg_pts * 1.05:
    print(f"  + Scoring: Score MORE vs {team_name} ({avg_pts_against:.1f} vs {avg_pts:.1f} avg) — favorable offensive matchup")
elif avg_pts_against < avg_pts * 0.95:
    print(f"  - Scoring: Score LESS vs {team_name} ({avg_pts_against:.1f} vs {avg_pts:.1f} avg) — {team_name} limits scoring well")
else:
    print(f"  = Scoring: About average vs {team_name} ({avg_pts_against:.1f} vs {avg_pts:.1f} avg)")
if avg_fg_pct_against > avg_fg_pct * 1.03:
    print(f"  + Shooting: Shoot MORE efficiently vs {team_name} ({avg_fg_pct_against:.3f} vs {avg_fg_pct:.3f} avg)")
elif avg_fg_pct_against < avg_fg_pct * 0.97:
    print(f"  - Shooting: Shoot LESS efficiently vs {team_name} ({avg_fg_pct_against:.3f} vs {avg_fg_pct:.3f} avg)")
else:
    print(f"  = Shooting: About average efficiency vs {team_name} ({avg_fg_pct_against:.3f} vs {avg_fg_pct:.3f} avg)")
if avg_tov_against < avg_tov * 0.95:
    print(f"  + Turnovers: Take BETTER care of the ball vs {team_name} ({avg_tov_against:.1f} vs {avg_tov:.1f} avg)")
elif avg_tov_against > avg_tov * 1.05:
    print(f"  - Turnovers: Turn it over MORE vs {team_name} ({avg_tov_against:.1f} vs {avg_tov:.1f} avg)")
else:
    print(f"  = Turnovers: About the same turnover rate vs {team_name} ({avg_tov_against:.1f} vs {avg_tov:.1f} avg)")

# --- DEFENSE ---
print("\nDEFENSE:")

if avg_stl_against > avg_stl * 1.05:
    print(f"  + Steals: More active hands vs {team_name} ({avg_stl_against:.1f} vs {avg_stl:.1f} avg) — force more turnovers")
elif avg_stl_against < avg_stl * 0.95:
    print(f"  - Steals: Fewer steals vs {team_name} ({avg_stl_against:.1f} vs {avg_stl:.1f} avg)")
else:
    print(f"  = Steals: About average vs {team_name} ({avg_stl_against:.1f} vs {avg_stl:.1f} avg)")

if avg_blk_against > avg_blk * 1.05:
    print(f"  + Blocks: Better rim protection vs {team_name} ({avg_blk_against:.1f} vs {avg_blk:.1f} avg)")
elif avg_blk_against < avg_blk * 0.95:
    print(f"  - Blocks: Weaker rim protection vs {team_name} ({avg_blk_against:.1f} vs {avg_blk:.1f} avg)")
else:
    print(f"  = Blocks: About average rim protection vs {team_name} ({avg_blk_against:.1f} vs {avg_blk:.1f} avg)")

# --- REBOUNDING ---
print("\nREBOUNDING:")

if avg_reb_against > avg_reb * 1.05:
    print(f"  + Rebounds: Out-rebound {team_name} ({avg_reb_against:.1f} vs {avg_reb:.1f} avg)")
elif avg_reb_against < avg_reb * 0.95:
    print(f"  - Rebounds: Get out-rebounded by {team_name} ({avg_reb_against:.1f} vs {avg_reb:.1f} avg)")
else:
    print(f"  = Rebounds: About even on the glass vs {team_name} ({avg_reb_against:.1f} vs {avg_reb:.1f} avg)")

if avg_oreb_against > avg_oreb * 1.05:
    print(f"  + Off. Rebounds: Get more second chances vs {team_name} ({avg_oreb_against:.1f} vs {avg_oreb:.1f} avg)")
elif avg_oreb_against < avg_oreb * 0.95:
    print(f"  - Off. Rebounds: Fewer second chances vs {team_name} ({avg_oreb_against:.1f} vs {avg_oreb:.1f} avg)")
else:
    print(f"  = Off. Rebounds: About average second chances vs {team_name} ({avg_oreb_against:.1f} vs {avg_oreb:.1f} avg)")

# --- OVERALL ---
print("\nOVERALL:")

if avg_pm_against > 0:
    print(f"  + Plus/Minus: Tend to WIN vs {team_name} (avg +/- {avg_pm_against:+.1f})")
elif avg_pm_against < 0:
    print(f"  - Plus/Minus: Tend to LOSE vs {team_name} (avg +/- {avg_pm_against:+.1f})")
else:
    print(f"  = Plus/Minus: Very close games vs {team_name} (avg +/- {avg_pm_against:+.1f})")

# Quick verdict
    
print("\nVERDICT:")
offense_edge = (avg_pts_against > avg_pts) and (avg_fg_pct_against > avg_fg_pct)
defense_edge = (avg_stl_against > avg_stl) or (avg_blk_against > avg_blk)

if offense_edge and defense_edge:
    print(f"  Strong all-around matchup — tend to perform well on both ends vs {team_name}.")
elif offense_edge:
    print(f"  Offensive edge vs {team_name}, but defense is a concern.")
elif defense_edge:
    print(f"  Play strong defense vs {team_name}, but scoring tends to dip.")
else:
    print(f"  Tough matchup vs {team_name} — tend to struggle on both ends.")
```
